```python
import matplotlib.pyplot as plt

labels = ['Nokia', 'Samsung', 'Apple', 'Lumia']
values = [20, 30, 45, 10]
colors = ['orange', '#b9d9c0', '#feb6ce', '#a7c4f6']
explode = (0, 0, 0.1, 0)

_, _, autotexts = plt.pie(values,labels=labels,colors=colors,explode=explode,shadow=True,startangle=-60,autopct='%1.1f%%')
plt.setp(autotexts, size=14, color='white')
plt.show()
```
