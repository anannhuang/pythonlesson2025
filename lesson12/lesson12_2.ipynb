```python
import matplotlib.pyplot as plt

labels = ['Nokia', 'Samsung', 'Apple', 'Lumia']
values = [20, 30, 45, 10]
colors = ['orange', 'green', 'red', 'blue']
explode = (0, 0, 0.1, 0)

_, _, autotexts = plt.pie(values,labels=labels,colors=colors,explode=explode,shadow=True,startangle=-65,autopct='%1.1f%%')
plt.setp(autotexts, size=14, color='white')
plt.show()
```
