### 設定隨機種子

設定 NumPy 隨機種子的方法是使用 `np.random.seed()`，並傳入一個整數作為種子值。這樣每次執行程式時，隨機數生成器都會從相同的起點開始，產生相同的隨機數序列。

範例如下：

```python
import numpy as np

np.random.seed(123)  # 設定隨機種子為123
scores = np.random.randint(50, 101, size=(50, 5))
print(scores)
```
總結：   
- 使用 `np.random.seed(整數)` 設定隨機種子
- 種子數值可自行設定，通常為任意整數
- 設定後，隨機數生成可重現，適合測試和除錯用途
